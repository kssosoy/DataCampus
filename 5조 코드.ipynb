```python
import numpy as np

# 진료과 코드 목록
unique_codes = data['DEPTT_NM'].unique()

# 각 진료과 코드 그룹을 고유한 값으로 정의한다.
g_codes = ['GA', 'GB', 'GC', 'GD', 'GE', 'GF', 'GG', 'GH', 'GJ', 'GK', 'GL', 'GN', 'GO', 'GQ', 'GR', 'GS', 'GT', 'GU', 'GV', 'GW', 'GX', 'GY', 'GZ']
h_m_codes = ['H1', 'HO', 'M1', 'M3', 'M5', 'M7', 'M8', 'M9', 'CM']

random_mapping = {code: np.random.randint(1, 10000) for code in set(unique_codes) - set(g_codes) - set(h_m_codes)}
g_value = np.random.randint(1, 10000)
h_m_value = np.random.randint(1, 10000)
random_mapping.update({code: g_value for code in g_codes})
random_mapping.update({code: h_m_value for code in h_m_codes})

print("임의의 정수 매핑:\n", random_mapping)

# 진료과 코드를 임의의 정수로 변환한다.
data['DEPTT_NM'] = data['DEPTT_NM'].map(random_mapping)
```

```python
def group_treatment_days(days):

    if days <= 10:
        return 10  # 10일 이내
    elif days <= 20:
        return 20
    elif days <=30:
        return 30
    elif days <=40:
        return 40
    elif days <=50:
        return 50
    elif days <=60:
        return 60
    elif days <=70:
        return 70
    elif days <=80:
        return 80
    elif days <=90:
        return 90
    elif days <=100:
        return 100
    elif days <=200:
        return 200
    elif days <=500:
        return 500
    elif days <= 1000:
        return 1000
    else :
      return 0
```

```python
data['ADM_YMD'] = pd.to_datetime(data['ADM_YMD'])
def date_to_season(date):
    month = date.month
    if 3 <= month <= 5:
        return 0  # 봄
    elif 6 <= month <= 8:
        return 1  # 여름
    elif 9 <= month <= 11:
        return 2  # 가을
    else:
        return 3  # 겨울

# 입원데이터를 이용하여 계절데이터 생성
data['입원시기'] = data['ADM_YMD'].apply(date_to_season)
```

```python
def gender_binary(gender_str):
    if gender_str == 'M':
        return 1
    elif gender_str == 'F':
        return 0
    else:
        return None 

# 성별 값을 바이너리형 변수로 변환
data['SEX_CD'] = data['SEX_CD'].apply(gender_binary)
data
```

```python
# 연령대별 중간값을 이용하여 처리
age_mapping = {
    '10대미만': 9 / 2,
    '10대': (10 + 19) / 2,
    '20대': (20 + 29) / 2,
    '30대': (30 + 39) / 2,
    '40대': (40 + 49) / 2,
    '50대': (50 + 59) / 2,
    '60대': (60 + 69) / 2,
    '70대': (70 + 79) / 2,
    '80대': (80 + 89) / 2,
    '90대': (90 + 99) / 2,
    '100대': (100 + 109) / 2,
    '100세 이상':100
}

data['AGRDE'] = data['AGRDE'].replace(age_mapping)

data
```

```python
# IQR 방식으로 이상치를 제거한다.
Q1 = data['치료기간'].quantile(0.25)
Q3 = data['치료기간'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 5 * IQR
upper_bound = Q3 + 5 * IQR

# 이상치를 포함한 행을 제거한다.
data = data[(data['치료기간'] >= lower_bound) & (data['치료기간'] <= upper_bound)]
data.info()
```

```python
def categorize_treatment_period(period):
    if 0 <= period <= 3:
        return 0  # 단기 치료기간 1~3일
    elif 4 <= period <= 7:
        return 1  # 중기 치료기간 4일~7일
    else:
        return 2 # 7일 초과

# 치료기간을 카테고리로 변환
data['치료기간'] = data['치료기간'].apply(categorize_treatment_period)
data['치료기간'].unique()
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import roc_curve, auc
from imblearn.combine import SMOTETomek
from itertools import cycle
from imblearn.over_sampling import SMOTE
import random
from imblearn.over_sampling import RandomOverSampler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score



X = data_without_null[['AGRDE', 'SEX_CD', 'DEPTT_NM',
       'WTHN30_REHOSP_YN','치료결과', 'ALT', 'Albumin', 'BUN', 'Glucose', 'WBC', 'AST',
       'Cholesterol', 'Creatinine', 'Protein', '맥박','CRP','MCH','MPV','Hemoglobin',
       '체온', '호흡', 'ICD10CD_A09.9','ICD10CD_J18.9 ',
       'ICD10CD_K35.8','medcount']]
y = data_without_null['입원기간']



# Oversampling
oversampler = RandomOverSampler(sampling_strategy={0:4000, 1: 4000,2: 4000}, random_state=24)
X_train_resampled, y_train_resampled = oversampler.fit_resample(X, y)
# com_sampler = SMOTETomek(sampling_strategy={0: 3000, 1: 3000, 2: 3000}, random_state=24)
# X_train_resampled, y_train_resampled = com_sampler.fit_resample(X_train, y_train)
# smote = SMOTE(sampling_strategy={1: 2500, 2: 2500}, random_state=24)
# X_train_resampled, y_train_resampled = smote.fit_resample(X, y)



random_5_numbers = random.sample(range(51,101), 1)
n_classes = 3
for q in random_5_numbers:
    X_train, X_test, y_train, y_test = train_test_split(X_train_resampled, y_train_resampled, test_size=0.2, random_state=q,stratify=y_train_resampled)





    # 모델 정의
    models = {
        '랜덤 포레스트': RandomForestClassifier(random_state=24),
        '그래디언트 부스팅': GradientBoostingClassifier(random_state=24),
        '로지스틱 회귀': LogisticRegression(),
        'XGBoost': XGBClassifier(random_state=24),
        'K-최근접 이웃': KNeighborsClassifier()
    }

    # 하이퍼파라미터 튜닝
    param_grids = {
        # '랜덤 포레스트': {'model__n_estimators': [50,100,150, 200], 'model__max_depth': [None, 10, 20], 'model__min_samples_split': [2,3,4,5]},
        '랜덤 포레스트': {'model__n_estimators': [150], 'model__max_depth': [20], 'model__min_samples_split': [2]},
        'XGBoost': {'model__n_estimators': [200], 'model__learning_rate': [0.1], 'model__max_depth': [8], 'model__gamma': [0.15], 'model__subsample': [0.8]},

        # 'XGBoost': {'model__n_estimators': [200,250,300], 'model__learning_rate': [0.1, 0.075], 'model__max_depth': [6, 7, 8], 'model__gamma': [0.05, 0.1, 0.15], 'model__subsample': [0.8]},
        # '그래디언트 부스팅': {'model__n_estimators': [350,400,450], 'model__learning_rate': [0.25,0.3]},
    }
    roc_dict = {}
    for model_name, model in models.items():
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('pca', PCA(n_components=15)),
            ('model', model)
        ])

        # pca = PCA(n_components=15)  # 2차원으로 축소
        # X_train_pca = pca.fit_transform(X_train)
        # X_test_pca = pca.transform(X_test)

        # GridSearchCV를 이용한 하이퍼파라미터 튜닝을 진행
        if model_name in param_grids:
            grid_search = GridSearchCV(pipeline, param_grids[model_name], scoring='accuracy', cv=5, n_jobs=-1)
            grid_search.fit(X_train, y_train)
            best_params = grid_search.best_params_
            pipeline.set_params(**best_params)
            print(f"{model_name} - 최적 하이퍼파라미터: {best_params}")


        pipeline.fit(X_train, y_train)

        train_preds = pipeline.predict(X_train)
        train_accuracy = accuracy_score(y_train, train_preds)


        test_preds = pipeline.predict(X_test)
        test_accuracy = accuracy_score(y_test, test_preds)

        print(f"{model_name} - Train Accuracy: {train_accuracy:.4f}, Test Accuracy: {test_accuracy:.4f}")

        if train_accuracy < test_accuracy:
            print(f"{model_name} - 과소적합 경향이 있습니다.")
        elif train_accuracy / test_accuracy > 1.1:
            print(f"{model_name} - 과대적합 경향이 있습니다.")
        else:
            print(f"{model_name} - 적절한 성능입니다.")

        # confusion matrix 출력
        cm = confusion_matrix(y_test, test_preds)
        print(f"{model_name} - Confusion Matrix:\n{cm}")

        # 실제 값과 예측 값 산점도 그리기
        plt.figure()
        sns.scatterplot(x=y_test, y=test_preds)
        plt.title(f'{model_name} Real-Predict')
        plt.xlabel('Real-value')
        plt.ylabel('predict-value')
        plt.show()
        report = classification_report(y_test, test_preds)
        print(f"{model_name} - Classification Report:\n{report}")

        if hasattr(model, 'predict_proba'):
            y_score = pipeline.predict_proba(X_test)
            fpr = dict()
            tpr = dict()
            roc_auc = dict()

            for i in range(n_classes):
                fpr[i], tpr[i], _ = roc_curve(y_test == i, y_score[:, i])
                roc_auc[i] = auc(fpr[i], tpr[i])

            roc_dict[model_name] = {'fpr': fpr, 'tpr': tpr, 'roc_auc': roc_auc}

    # 전체 모델 ROC curve 합치기
    plt.figure(figsize=(7, 7))

    for model_name, model_roc_params in roc_dict.items():
        fprs = model_roc_params["fpr"]
        tprs = model_roc_params["tpr"]
        rocs = model_roc_params["roc_auc"]

        all_fpr = np.unique(np.concatenate([fprs[i] for i in range(n_classes)]))
        mean_tpr = np.zeros_like(all_fpr)

        for i in range(n_classes):
            mean_tpr += np.interp(all_fpr, fprs[i], tprs[i])

        mean_tpr /= n_classes
        mean_roc_auc = auc(all_fpr, mean_tpr)

        plt.plot(all_fpr, mean_tpr, label=f"{model_name} (avg. AUC={mean_roc_auc:.2f})")

    plt.plot([0, 1], [0, 1], "k--")
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve for All Models")
    plt.legend(loc="lower right")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# 모델의 성능을 학습 곡선으로 시각화
def plot_learning_curve(model, X, y):
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10))
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.figure(figsize=(10, 6))
    plt.title("Learning Curve")
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.grid()
    
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std*0.5, train_scores_mean + train_scores_std*0.5, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std*0.5, test_scores_mean + test_scores_std*0.5, alpha=0.1, color="g")
    
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    
    plt.legend(loc="best")
    plt.show()

# 각 모델에 대해 학습 곡선 그리기
for model_name, model in models.items():
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('pca', PCA(n_components=15)),
        ('model', model)
    ])
    plot_learning_curve(pipeline, X_train_resampled, y_train_resampled)
```
